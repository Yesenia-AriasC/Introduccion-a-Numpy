# CONCATENACIÓN

## [.concatenate()](https://www.analyticslane.com/2021/08/17/numpy-concatenar-matrices-en-numpy-con-np-concatenate/)

Ambos arreglos __tienen que tener la misma dimensión__.

* __axis:__ eje a lo largo del cual se desea que se concatenan los vectores o matrices. Los posibles valores son
* __0:__ las matrices se unen por filas, siendo este el __valor por defecto__ de la función.
* __1:__ las matrices se unen por columnas.
None: al seleccionar esta opción todas las matrices se aplanan y se concatenan como vectores.

__Axis = 0__ hace referencia a las _filas_, mientras que __axis = 1__ a las _columnas_:

IMG_260E3E0A57A4-1.jpeg

### Ejemplo 1

````python
a = np.array([[1,2], [3,4]])
b = np.array([5, 6])
````

````python
# Tienen dimensiones diferentes
print(a.ndim)
print(b.ndim)
````

````python
#error
#np.concatenate((a,b), axis =0)
````

Expandimos b a 2 dimensiones para poder concatenar:


__Axis = 0__ hace referencia a las _filas_, mientras que __axis = 1__ a las _columnas_.

````python
# Aumenta dimensión de b para poder concatenar
b = np.expand_dims(b, axis = 0)
print(b.ndim,a.ndim)
print(a,end="\n\n")
print(b)
````

````python
#Concatena colocando la matriz a sobre la matriz b
np.concatenate((a,b), axis = 0)
````

###Ejemplo 2

concatenemos de forma diferente


__Axis = 0__ hace referencia a las _filas_, mientras que __axis = 1__ a las _columnas_.

````python
a = np.array([[1,2], [3,4]])
b = np.array([5, 6])
````

````python
a2 = a.reshape(4,1)
print(a2, a2.ndim, a2.size, sep="\n")# array y dimension
````

````python
b2 = np.expand_dims(b, axis = 1)
print(b2, b2.ndim, b2.size , sep="\n")# array y dimension
````

````python
#Concatena colocando la matriz a2 sobre la matriz b2
np.concatenate((a2,b2), axis = 0)
````
