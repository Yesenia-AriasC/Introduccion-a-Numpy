Hay 2 funciones muy importantes de los arreglos (Shape y Reshape). 

La forma de un arreglo nos va a decir con que estructura se está trabajando (tamaño, manipular, ingresar).

Material [aquí](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.shape.html).

# Shape
Forma del arreglo.

```python
arr = np.random.randint(1,10,(3,2))
print(arr, end="\n\n")
arr.shape
```

# Reshape
Transforma el arreglo __mientras se mantengan los elementos__.

Tiene que tomarse en cuenta los elementos, es decir cambiar el shape respetando el número de elementos del array inicial.

```python
arr.reshape(1,6)
```

```python
arr.reshape(1,6).shape
```

```python
arr.reshape(2,3)
```

segunda forma de hacerlo

```python
np.reshape(arr,(1,6))
```

## Reshape como lo haría __C__

Apila los valores por filas.

image.png

Creemos un array para ver cómo funciona este reshape y el siguiente:

```python
arr2 = np.arange(1,7).reshape(3,2)        # Creo un array con los números del 1 al 6 y que tenga shape (3,2)
arr2
```

Miremos cómo funciona este reshape:

```python
np.reshape(arr2,(2,3), 'C')
```

## Reshape como lo haría __Fortran__

Apila los valores por columnas.

image.png

```python
np.reshape(arr2,(2,3), 'F')
```

## Como esté optimizado el computador

En este caso es como en __C__.
